```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("D:/Prodigy Infotech/Data Science/Task 1/6. India Census 2011.csv")
```

```python
df.head()
```

```python
population_data=df[['State_name','Population','Male','Female']]
```

```python
plt.figure(figsize=(10,12))
plt.barh(population_data['State_name'], population_data['Population'])
plt.xlabel('Population')
plt.ylabel('States')
plt.title('Population Distribution by State')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.histh([population_data['Male'], population_data['Female']], bins=10, 
         color=['blue', 'pink'], edgecolor='black', label=['Male', 'Female'])
plt.xlabel('Population')
plt.ylabel('Number of States')
plt.title('Population Distribution by Gender')
plt.legend()
plt.show()
```
